```python
import numpy as np
x1=[10,8,15]
x2=[8,20,25]
y=[20,15,30]
```

### Find the best value of w1,w2,b that fits the following equation


```python
np.random.seed(123)
random_init=np.random.randn(3)
w1,w2,b=random_init[0],random_init[1],random_init[2]
```

```python
print(f"Initialized weights -w1 :{w1},w2:{w2},b:{b}" )
```

```python
n=len(x1)
def forward_pass(w1,w2,b):
    y_pred=[]
    total_error=0
    for i in range(n):
        y_hat=w1*x1[i]+w2*x2[i]+b
        j=(y_hat-y[i])**2
        total_error=total_error+j
        y_pred.append(y_hat)
    return total_error,y_pred
                             
```

```python
total_error,y_pred=forward_pass(w1,w2,b)
```

```python
lr=0.1
def weight_update(w1,w2,b,y_pred):
    for i in range(n):
        dj_dw1=2*(y_pred[i]-y[i])*x1[i]
        dj_dw2=2*(y_pred[i]-y[i])*x2[i]
        dj_b=2*(y_pred[i]-y[i])
        w1=w1-lr*dj_dw1
        w2=w2-lr*dj_dw2
        b=b-lr*dj_b
    return w1,w2,b     
    
    
    

    
     
    
```

### Model Orchestration

```python
n=len(x1)
lr=0.05
iterations=100
#random initialization of weights and bias
cycles=5
cnt=0
j=0
while cnt==0 and j<cycles :
    error=[]
    np.random.seed(123)
    random_init=np.random.randn(3)
    w1,w2,b=random_init[0],random_init[1],random_init[2]
    
    print(f"Learning rate {lr}")
    for i in range(iterations):
        
        total_error,ypred=forward_pass(w1,w2,b)
        error.append(total_error)
        print(f" total error {total_error} ,iterations {i}")
        if total_error<20:
            print("early stopping as the condition is met")
            print(f"w1: {w1} w2: {w2} bias:{b}")
            cnt=1
            break
            
        else:
            w1,w2,b=weight_update(w1,w2,b,ypred)
    lr=lr/10
    print()
    print("""   **********  """)
    
    print(f"Completed Cycle {j}")
    j=j+1
    print()      
print("training completed")    

    
          
    
    

          
          

```
